```python
#importar las librerias a ocupar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#importar los datos
df = pd.read_csv('https://raw.githubusercontent.com/Criztian02/Reto/main/Libro1.csv')
df
```

```python
#graficar de los datos
plt.figure(figsize=(12,6))
plt.plot(df['Clave Indicador'],df['2015'],'-o',color='blue',label='2015')
plt.plot(df['2016'],'-o',color='red',label='2016')
plt.plot(df['2017'],'-o',color='green',label='2017')
plt.plot(df['2018'],'-o',color='yellow',label='2018')
plt.plot(df['2019'],'-o',color='orange',label='2019')
plt.legend(loc='best')
plt.show()
```

```python
#pronostico del 2020
from statsmodels.tsa.api import SimpleExpSmoothing
mod = SimpleExpSmoothing(df['2019']).fit(smoothing_level=0.3, optimized=False)
df['2020']=mod.fittedvalues
df
```

```python
#grafica de los datos 2019-2020
plt.figure(figsize=(12,6))
plt.plot(df['Clave Indicador'],df['2019'],'-o',color='blue',label='2019')
plt.plot(df['2020'],'-o',color='red',label='2020')
plt.legend(loc='best')
plt.show()
```
